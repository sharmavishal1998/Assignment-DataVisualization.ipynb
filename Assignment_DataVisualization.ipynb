```python
#Question9- Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.
#Answer-
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate a dataset
np.random.seed(42)  # For reproducibility
categories = ['A', 'B', 'C', 'D']
values = []

# Create random values for each category
for category in categories:
    values.extend(np.random.normal(loc=10 if category == 'A' else 20 if category == 'B' else 30 if category == 'C' else 40,
                                   scale=5,
                                   size=100))  # 100 random values for each category

# Create a DataFrame
df = pd.DataFrame({
    'Category': np.repeat(categories, 100),  # Repeat categories
    'Value': values
})

# Create box plot to visualize distribution
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Value', data=df)

# Add titles and labels
plt.title("Distribution of Values Across Different Categories")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()

```

```python
#Question10-Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.
#Answer-
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Generate synthetic dataset
np.random.seed(42)  # For reproducibility
num_samples = 100

# Generate correlated features
x1 = np.random.rand(num_samples)
x2 = x1 * 0.5 + np.random.rand(num_samples) * 0.5  # Correlated with x1
x3 = x1 * -0.3 + np.random.rand(num_samples) * 0.5  # Inversely correlated with x1
x4 = np.random.rand(num_samples)  # Uncorrelated

# Create DataFrame
df = pd.DataFrame({
    'Feature 1': x1,
    'Feature 2': x2,
    'Feature 3': x3,
    'Feature 4': x4
})

#Calculate the correlation matrix
correlation_matrix = df.corr()

#Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True)

# Add titles and labels
plt.title("Correlation Matrix Heatmap")
plt.show()

```

```python
#Question14- Using the given x and y data, generate a 3D surface plot to visualize the function
"""x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)"""
#Answer-
import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Generate x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Compute z values
z = np.sin(np.sqrt(x**2 + y**2))

# Create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create a 3D surface plot using Plotly
fig = go.Figure(data=[go.Surface(z=df['Z'].values.reshape(100, 100),
                                   x=df['X'].values.reshape(100, 100),
                                   y=df['Y'].values.reshape(100, 100))])

# Update layout
fig.update_layout(title='3D Surface Plot of z = sin(sqrt(x^2 + y^2))',
                  scene=dict(xaxis_title='X',
                             yaxis_title='Y',
                             zaxis_title='Z'),
                  autosize=True)
fig.show()

```

```python
#Question16-Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
#Answer-
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Set up the output to display in a Jupyter Notebook
output_notebook()

# Generate x values and compute y values (sine of x)
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a Bokeh figure
p = figure(title="Sine Wave", x_axis_label='X', y_axis_label='sin(X)',
           width=800, height=400)

# Add a line renderer for the sine wave
p.line(x, y, legend_label="sin(x)", line_color="blue", line_width=2)
show(p)
```

```python
#Question17-Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.
#Answer-
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Set up the output to display in a Jupyter Notebook
output_notebook()

# Generate random data
np.random.seed(42)  # For reproducibility
n = 50  # Number of points
x = np.random.rand(n) * 100  # Random x values
y = np.random.rand(n) * 100  # Random y values
sizes = np.random.randint(10, 50, size=n)  # Random sizes for markers
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], size=n)  # Random colors

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Create a Bokeh figure
p = figure(title="Scatter Plot with Random Sizes and Colors",
           x_axis_label='X', y_axis_label='Y',
           width=800, height=400)

# Add a scatter renderer using ColumnDataSource
p.scatter('x', 'y', size='sizes', color='colors', source=source, fill_alpha=0.6)

# Show the plot
show(p)




```

```python
#Question18-Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
"""fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]"""
#Answer-
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Set up the output to display in a Jupyter Notebook
output_notebook()

# Define the dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a Bokeh figure
p = figure(x_range=fruits, title="Fruit Counts",
           x_axis_label='Fruits', y_axis_label='Counts',
           width=800, height=400)

# Add a vbar (vertical bar) renderer
p.vbar(x=fruits, top=counts, width=0.5, color='blue')

# Show the plot
show(p)
```

```python
#Question19-Create a Bokeh histogram to visualize the distribution of the given data.
"""data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)"""
# Answer-
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Set up the output to display in a Jupyter Notebook
output_notebook()

# Generate random data
data_hist = np.random.randn(1000)

# Compute the histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh figure
p = figure(title="Histogram of Random Data",
           x_axis_label='Value',
           y_axis_label='Count',
           width=800, height=400)

# Add a quad renderer to create the histogram
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color='blue', line_color='black')
show(p)
```

```python
#Question20-Create a Bokeh heatmap using the provided dataset.
"""data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)"""
#Answer-
from bokeh.plotting import figure, show
from bokeh.models import LinearColorMapper, ColorBar
from bokeh.io import output_notebook
import numpy as np

# Enable output in Jupyter Notebook
output_notebook()

# Generate data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)

# Create a color mapper
mapper = LinearColorMapper(palette="Viridis256", low=data_heatmap.min(), high=data_heatmap.max())

# Create a figure
p = figure(title="Heatmap Example", x_range=(0, 1), y_range=(0, 1), toolbar_location=None)

# Add image glyph
p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=mapper)

# Add a color bar
color_bar = ColorBar(color_mapper=mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
